```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, accuracy_score
from sklearn.impute import SimpleImputer
```

```python
train_df = pd.read_csv("/kaggle/input/datarush-2023/train.csv")
test_df = pd.read_csv('/kaggle/input/datarush-2023/test.csv')
train_df.dropna(axis=0,inplace=True)
test_df.dropna(axis=0,inplace=True)
test_df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(train_df,train_df.category)
x_train.shape,y_train.shape
```

```python
tfidf = TfidfVectorizer(sublinear_tf= True, 
                       min_df = 5, 
                       norm= 'l2', 
                       ngram_range= (1,2), 
                       stop_words ='english')


```

```python
X_train_counts = tfidf.fit_transform(x_train.abstract) #train+transform 
# X_test_counts = tfidf.transform(x_test.abstract) #transform

```

```python
clf = MultinomialNB()
clf.fit(X_train_counts, y_train) # w,b 

```

```python

X_test_counts = tfidf.transform(test_df.abstract)
pred = clf.predict(X_test_counts)
labels = pred
solution = test_df[['id']].copy()
solution['category'] = labels
solution.to_csv('submission.csv', encoding='utf-8', index=False)


```
