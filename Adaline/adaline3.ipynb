```python
import numpy as np

# Define the inputs and target output for the AND function
X = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
y = np.array([-1, -1, -1, 1])

# Initialize the weights and bias
w = np.array([0.1, 0.2])
b = 0

# Define the Adaline algorithm
def adaline_AND(X, y, w, b, learning_rate, epochs):
    for epoch in range(epochs):
        for i in range(len(X)):
            # Compute the net input
            net_input = np.dot(X[i], w) + b

            # Compute the output using the step function
            output = np.where(net_input > 0, 1, -1)

            # Update the weights and bias
            error = y[i] - output
            w += learning_rate * error * X[i]
            b += learning_rate * error

    # Return the weights and bias
    return w, b

# Train the model and get the weights and bias
w, b = adaline_AND(X, y, w, b, learning_rate=0.1, epochs=100000)

# Define a function to make predictions
def predict_AND(x, w, b):
    net_input = np.dot(x, w) + b
    output = np.where(net_input > 0, 1, -1)
    return output

# Test the model by making predictions for all inputs
for i in range(len(X)):
    prediction = predict_AND(X[i], w, b)
    print(f"Input: {X[i]}, Predicted output: {prediction}")
```
